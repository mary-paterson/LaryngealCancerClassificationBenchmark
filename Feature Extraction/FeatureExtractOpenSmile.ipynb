```python
import opensmile
import pandas as pd
import os
```

```python
cwd = os.getcwd()
parent_dir = os.path.dirname(cwd)
```

```python
source_dir = f'{parent_dir}/Audio/'
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
```

```python
feats = smile.process_folder(source_dir)
```

```python
feats.head()
```

```python
def format_features(df):
    df = df.reset_index()
    df = df.rename({'file': 'filename'}, axis=1)
    df['file'] = df['filename'].str.split('\\').str[-1]
    df = df.drop(['start', 'end', 'filename'], axis=1)

    return df
```

```python
feats_formatted = format_features(feats)
feats_formatted.head()
```

```python
feats_formatted.to_csv(f'{parent_dir}/Audio Features/OpenSmile_FEMH.csv')
```
